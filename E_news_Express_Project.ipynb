1. Do the users spend more time on the new landing page than the old landing page?
Perform visual analysis

```python
sns.boxplot(data = df, x = 'landing_page', y = 'time_spent_on_the_page', palette=['blue', 'green'])
```

 Formulate null and alternative hypotheses

```python
df.groupby(['landing_page'])['time_spent_on_the_page'].mean()
```

Step 1: Define null and alternate hypotheses

1.  H0 - The mean time spent on the old landing page.

2. H1 - The mean time spent on the new landing page.
Let  μ1, μ2
 be the mean time spent on the old landing page vs the mean time spent on the new landing page.

Mathematically, the above formulated hypotheses can be written as:
H0 =  μ1= μ2
H1 =  μ1< μ2
Note: As given in the problem statement, we select
 alpha = 0.05.





Select the appropriate test
This is a case of a one-tailed two-sample mean test. As the two samples are not independent in this case, and also the population standard deviations are unknown, we shall use a paired t-test for this problem.

Calculate the p-value

```python
mu1 = df[df['landing_page']=='old']['time_spent_on_the_page']
mu2 = df[df['landing_page']=='new']['time_spent_on_the_page']
```

```python
# find the p-value
test_stat, p_value  = stats.ttest_rel(mu2, mu1, alternative = 'greater')  #Performing a paired t-test

print('The p-value is', p_value)
```

```python
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

Write the inference based on the p-value

As given Alpha -0.05(5%) At 5% significance level, we reject the null hypothesis. Hence we have enough statistical evidence to say that the new landing page encourages users to spend more time than on the old landing page.

the p value is 0.0003 which is less than the 0.05. Hence, we reject the null hypothesis.

This means that the new landing page encourages users to spend more time significantly.

Scenario 5:

Is the mean time spent on the new page same for the different language users?


- Perform visual analysis - Formulate null and alternative hypotheses - Select the appropriate test - Calculate the p-value - Write the inference based on the p-value

```python
df1 = df.loc[df['landing_page'] == 'new']
df1
```

Total analysis values are shown along with graphs.

```python
df1.groupby(['language_preferred'])['time_spent_on_the_page'].mean()
```

calculated mean

```python
sns.boxplot(data = df1, x = 'language_preferred', y = 'time_spent_on_the_page',palette=['blue','green'])
```

plotted the graph
The average time spent on the pages seems relatively the same among the different language preferences.

Let's write the null and alternative hypothesis

μ1,μ2,μ3be the means of time spent on the new landing page for language preferences English, Spanish and French respectively.

null hypothesis
H0 = μ1=μ2=μ3

alternative hypothesis
Hα =  At least one time average of the new landing page is different among the language preferences.

```python
from scipy.stats import f_oneway

# perform one-way anova test
test_stat, p_value = f_oneway(df1.loc[df1['language_preferred'] == 'English', 'time_spent_on_the_page'],
                              df1.loc[df1['language_preferred'] == 'Spanish', 'time_spent_on_the_page'],
                              df1.loc[df1['language_preferred'] == 'French', 'time_spent_on_the_page'])
print('The p-value is ' + str(p_value))
```

Compare the p-value with α

```python
# print the conclusion based on p-value
if p_value < 0.05:
    print(f'As the p-value {p_value} is less than the level of significance, we reject the null hypothesis.')
else:
    print(f'As the p-value {p_value} is greater than the level of significance, we fail to reject the null hypothesis.')
```

In this scenario, the p value is 0.432 which is greater than the 0.05. Hence, we fail to reject the null hypothesis.

This means that the time averages on the new landing page for the different languages are relatively the same.
